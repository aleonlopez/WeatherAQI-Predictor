```python
# Drop 'Unnamed' column
weather_train_df = weather_train_df.drop(['Unnamed: 0'], axis=1)
```

```python
import numpy as np

# Gets the day, month, and year from the 'Date' column
def time_split(df):
    df['Year'] = df['Date'].str.slice(0, 4).astype(int)
    df['Month'] = df['Date'].str.slice(5, 7).astype(int)
    df['Day'] = df['Date'].str.slice(8, 10).astype(int)

time_split(weather_train_df)
```

```python
# Convert month and day into sin and cos
def feature_sin_cos(df):
    df['Month_sin'] = np.sin(df.Month * 2 * np.pi / 12)
    df['Month_cos'] = np.cos(df.Month * 2 * np.pi / 12)

    df['Day_sin'] = np.sin(df.Day * 2 * np.pi / 31)
    df['Day_cos'] = np.cos(df.Day * 2 * np.pi / 31)

feature_sin_cos(weather_train_df)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Month sin
axs[0, 0].scatter(weather_train_df['Month'], weather_train_df['Month_sin'])
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Sin of month')
axs[0, 0].set_title('Sin of month (for feature engineering)')

# Month cos
axs[0, 1].scatter(weather_train_df['Month'], weather_train_df['Month_cos'])
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Cos of month')
axs[0, 1].set_title('Cos of month (for feature engineering)')

# Day sin
axs[1, 0].scatter(weather_train_df['Day'], weather_train_df['Day_sin'])
axs[1, 0].set_xlabel('Day')
axs[1, 0].set_ylabel('Sin of day')
axs[1, 0].set_title('Sin of day (for feature engineering)')

# Day cos
axs[1, 1].scatter(weather_train_df['Day'], weather_train_df['Day_cos'])
axs[1, 1].set_xlabel('Day')
axs[1, 1].set_ylabel('Cos of day')
axs[1, 1].set_title('Cos of day (for feature engineering)')

plt.tight_layout()
plt.show()
```

```python
# Drop 'Day', 'Month', columns
weather_train_df = weather_train_df.drop(['Day'], axis=1)
weather_train_df = weather_train_df.drop(['Month'], axis=1)
```

```python
# Encoding columns from str to int

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

# Events
weather_train_df['Events'] = label_encoder.fit_transform(weather_train_df['Events'].astype(str))
weather_train_df['Events'] = weather_train_df['Events'].astype(int)

# Precipitation
weather_train_df['PrecipitationIn'] = label_encoder.fit_transform(weather_train_df['PrecipitationIn'].astype(str))
weather_train_df['PrecipitationIn'] = weather_train_df['PrecipitationIn'].astype(int)

# WindDirDegrees
weather_train_df['WindDirDegrees.br...'] = label_encoder.fit_transform(weather_train_df['WindDirDegrees.br...'].astype(str))
weather_train_df['WindDirDegrees.br...'] = weather_train_df['WindDirDegrees.br...'].astype(int)

# Season
weather_train_df['season'] = label_encoder.fit_transform(weather_train_df['season'].astype(str))
weather_train_df['season'] = weather_train_df['season'].astype(int)

weather_train_df
```
